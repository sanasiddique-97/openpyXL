```python
import pandas as pd
```

```python
tuples = [
         ("", "","Student No."),
    ("", "","Description"),
    ("", "","Pass / Fail"),
    ("", "","Remarks"),
    ("", "","Grade"),
    ("", "","Root Cause"),
    ("", "","Exam."),
    ("", "","Exam Date"),
    ("", "","Practical"),
    ("", "","Practical Date"),
         ('Vehicle MOB','2021','Nov'),
         ('Vehicle MOB','2021','Dec'),
         ('Vehicle MOB','2022','Jan'),
         ('Vehicle MOB','2022','Feb'),
         ('Vehicle MOB','2022','Mar'),
         ('Vehicle MOB','2022','Apr'),
         ('Vehicle MOB','2022','May'),
         ('Vehicle MOB','2022','Jun'),
         ('Vehicle MOB','2022','Jul'),
         ('Vehicle MOB','2022','Aug'),
         ('Vehicle MOB','2022','Sep'),
         ('Vehicle MOB','2022','Oct'),
         ('Vehicle MOB','2022','Nov'),
         ('Vehicle MOB','2022','Dec'),
         ('Vehicle MOB','2022MY' ,'Total')]
Final=pd.MultiIndex.from_tuples(tuples)
df = pd.DataFrame(index=range(16), columns=Final)
```

```python
# df3.to_csv('out3.csv', index = False)
```

```python
df['Vehicle MOB']['2021']['Nov'][15] = df['Vehicle MOB']['2021']['Nov'].sum(axis = 0)
```

```python
df['Vehicle MOB']['2021']['Nov'][15]
```

```python
df['Vehicle MOB']['2021']['Dec'][15] = df['Vehicle MOB']['2021']['Dec'].sum(axis = 0)
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
for i in months:
    df['Vehicle MOB']['2022'][i][15] = df['Vehicle MOB']['2022'][i].sum(axis = 0)   
```

```python
sum_of_2021 =df['Vehicle MOB']['2021']['Nov'] + df['Vehicle MOB']['2021']['Dec']
```

```python
sum_2022 = 0
for i in months:
    
     sum_2022 += df['Vehicle MOB']['2022'][i]
    
```

```python
df['Vehicle MOB']['2022MY']["Total"] = sum_of_2021 + sum_2022
```

```python
df['Vehicle MOB']['2022MY']["Total"][15] = df['Vehicle MOB']['2022MY']["Total"].sum(axis = 0)
```

```python
df[""][""]["Practical Date"][15] = "Total Presents"
```

```python
from openpyxl.styles.borders import Border,Side
import openpyxl
from openpyxl import Workbook
from openpyxl.styles import Color, PatternFill, Font, Border
from openpyxl.styles import colors
from openpyxl.cell import Cell
```

```python
wb=openpyxl.load_workbook('final_output.xlsx')
sheet = wb.sheetnames
worksheet= wb['Sheet1']
from openpyxl.styles import Alignment
worksheet.merge_cells("A18:I18")#merging Last row cells
from openpyxl.formatting.rule import CellIsRule
# Conditional Formating
gfill = PatternFill(start_color='3CB371',end_color='3CB371',fill_type='solid')
worksheet.conditional_formatting.add('K18:X18',CellIsRule(operator='equal',formula=['0'],stopIfTrue=True,fill=gfill))
rfill = PatternFill(start_color='FF6666',end_color='FF6666',fill_type='solid')
worksheet.conditional_formatting.add('K18:X18',CellIsRule(operator='greaterThan',formula=['8'],stopIfTrue=True,fill=rfill))
ofill = PatternFill(start_color='FFB266',end_color='FFB266',fill_type='solid')
worksheet.conditional_formatting.add('K18:X18',CellIsRule(operator='between',formula=('1','8'),stopIfTrue=True,fill=ofill))
yfill = PatternFill(start_color='FFB266',end_color='FFB266',fill_type='solid')
worksheet.conditional_formatting.add('K18:X18',CellIsRule(operator='lessThan',formula=['2'],stopIfTrue=True,fill=yfill))
wb.save('output199431decemmmmmnn.xlsx')
```
